```python
import pandas as pd
from sklearn.preprocessing import scale

raw1 = pd.read_excel('experiment8.xlsx')                  #read both sets of cutter data inside
raw2 = pd.read_excel('experiment8_1.xlsx')
```

```python
print(raw1.shape)               #shows record, column numbers for both datasets
print(raw2.shape) 
```

```python
#checks for null values. The first .sum() counts null for each column, the scond .sum() adds them up
print(raw1.isnull().sum().sum())
print(raw2.isnull().sum().sum())
```

```python
print(raw1[raw1['fca'] != raw1['fa']])  #show records where fca is different from fa.
print(raw2[raw2['fca'] != raw2['fa']])  #both return empty DF -> fca = fa for both CSV files

raw1d = raw1.drop('fca', axis = 1)      #drops fca to reduce parameter count
raw2d = raw2.drop('fca', axis = 1)      #raw1d(ropped), raw2d(ropped)
```

```python
raw1ds = scale(raw1d)                   #standardizing the data from raw1d so that mean = 0 , std = 1
raw2ds = scale(raw2d)                   

raw1ds = pd.DataFrame(raw1ds, columns = raw1d.columns)     #scale() produces array. Converting array to DF exporting
raw2ds = pd.DataFrame(raw2ds, columns = raw2d.columns)

raw1ds['measured_wearout'] = raw1d['measured_wearout']    #However, we want to keep the original measured_wearout values
raw2ds['measured_wearout'] = raw2d['measured_wearout']
```

```python
raw1d.describe()        #before scaling
```

```python
raw1ds.describe()      #after scaling. Note that measured_wearout remained the same while other parameters are scaled
```

```python
raw1ds.to_excel('experiment8a.xlsx', index = False)           #exports preprocessed datasets to Downloads folder
raw2ds.to_excel('experiment8a_1.xlsx', index = False)
```
